```python
def get_chapter(lemme_to_find) :
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme==lemme_to_find :
                return chapter
            
def th_string_to_pred(th_string,seuil=0.05,model=model,tokenizer=t,max_length=max_length) :
    x = th_declaration_to_text(th_string)
    x = pretraitement(np.array([x]))
    
    encoded_docs = tokenizer.texts_to_sequences(x)
    
    padded_docs = prepro.sequence.pad_sequences(encoded_docs, maxlen=max_length, padding='post')
    
    y_pred = model.predict(padded_docs)[0]
    max_pred_value = np.max(y_pred)
    y_pred = np.array([0 if val < max_pred_value*seuil else 1 for val in y_pred])
    
    return y_pred
        
def lemme_pred(lemme,seuil=0.05,model=model,tokenizer=t,max_length=max_length) :
    x = th_declaration_to_text(all_th_declaration[lemme])
    x = pretraitement(np.array([x]))
    
    encoded_docs = tokenizer.texts_to_sequences(x)
    
    padded_docs = prepro.sequence.pad_sequences(encoded_docs, maxlen=max_length, padding='post')
    
    y_pred = model.predict(padded_docs)[0]
    max_pred_value = np.max(y_pred)
    y_pred = np.array([0 if val < max_pred_value*seuil else 1 for val in y_pred])
    
    return y_pred

def th_string_to_pred_top_k(th_string,k=5,model=model,tokenizer=t,max_length=max_length) :
    x = th_declaration_to_text(th_string)
    x = pretraitement(np.array([x]))
    
    encoded_docs = tokenizer.texts_to_sequences(x)
    
    padded_docs = prepro.sequence.pad_sequences(encoded_docs, maxlen=max_length, padding='post')
    
    y_pred = model.predict(padded_docs)[0]
    pred_sort = sorted(y_pred)[::-1]
    index_top_k=[list(y_pred).index(pred_sort[i]) for i in range(k)]
    new_y_pred = np.zeros(len(y_pred))
    for i in range(k) :
        new_y_pred[index_top_k[i]]=1
        
    return new_y_pred

def lemme_pred_top_k(lemme,k=5,model=model,tokenizer=t,max_length=max_length) :
    x = th_declaration_to_text(all_th_declaration[lemme])
    x = pretraitement(np.array([x]))
    
    encoded_docs = tokenizer.texts_to_sequences(x)
    
    padded_docs = prepro.sequence.pad_sequences(encoded_docs, maxlen=max_length, padding='post')
    
    y_pred = model.predict(padded_docs)[0]
    pred_sort = sorted(y_pred)[::-1]
    index_top_k=[list(y_pred).index(pred_sort[i]) for i in range(k)]
    new_y_pred = np.zeros(len(y_pred))
    for i in range(k) :
        new_y_pred[index_top_k[i]]=1
        
    return new_y_pred

def is_good_pred(lemme,seuil=0.05,model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    pred_y = lemme_pred(lemme,seuil,model,tokenizer)
    
    pred_y_ones = np.where(pred_y==1)[0]
    
    
    labeled_y = dep_to_list(dep_dir_path+"/"+get_chapter(lemme)+"/"+lemme+".dep")
    y = multilabel_binarizer.transform(np.array([labeled_y]))
    
    y_ones = np.where(y[0]==1)[0]
    
    compare = [1 if val in pred_y_ones else 0 for val in y_ones]
    
    if len(compare)==0 :
        compare=[0]
    
    return (np.average(compare),len(pred_y_ones)/len(pred_y))

def is_good_pred_top_k(lemme,k=5,model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    pred_y = lemme_pred_top_k(lemme,k,model,tokenizer)
    
    pred_y_ones = np.where(pred_y==1)[0]
    
    
    labeled_y = dep_to_list(dep_dir_path+"/"+get_chapter(lemme)+"/"+lemme+".dep")
    y = multilabel_binarizer.transform(np.array([labeled_y]))
    
    y_ones = np.where(y[0]==1)[0]
    
    compare = [1 if val in pred_y_ones else 0 for val in y_ones]
    
    if len(compare)==0 :
        compare=[0]
    
    return (np.average(compare),len(pred_y_ones)/len(pred_y))


print(is_good_pred_top_k('cong3_bet_eq',k=20))
```

```python
def test_pred(seuil=0.05,model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    sum_res=0
    size=0
    fail_size=0
    total_precision=None
    total_prop_classe=None
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme in th_valid :
                (res,prop_classe) = is_good_pred(lemme,seuil,model,tokenizer,multilabel_binarizer)
                if res==1 :
                    sum_res+=1
                else :
                    if total_precision is None :
                        total_precision=res
                    else :
                        total_precision+=res
                    fail_size+=1
                if total_prop_classe is None :
                    total_prop_classe=prop_classe
                else :
                    total_prop_classe+=prop_classe
                size+=1
    if total_precision is None :
        total_precision=0
    if total_prop_classe is None :
        total_prop_classe=0
    return (sum_res/size,total_prop_classe/size,total_precision/fail_size)

def test_pred_top_k(k=5,model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    sum_res=0
    size=0
    fail_size=0
    total_precision=None
    total_prop_classe=None
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme in th_valid :
                (res,prop_classe) = is_good_pred_top_k(lemme,k,model,tokenizer,multilabel_binarizer)
                if res==1 :
                    sum_res+=1
                else :
                    if total_precision is None :
                        total_precision=res
                    else :
                        total_precision+=res
                    fail_size+=1
                if total_prop_classe is None :
                    total_prop_classe=prop_classe
                else :
                    total_prop_classe+=prop_classe
                size+=1
    if total_precision is None :
        total_precision=0
    if total_prop_classe is None :
        total_prop_classe=0
    return (sum_res/size,total_prop_classe/size,total_precision/fail_size)
```

```python
def find_seuil(model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    seuil_table=[]
    acc_table=[]
    prop_table=[]
    prev_seuil=0
    seuil=0.5
    accuracy=0
    prop_classe=0.5
    while(accuracy < 1 and prop_classe < 1 and abs(seuil-prev_seuil) > 1e-6) :
        (accuracy,prop_classe,precision) = test_pred(seuil,model,tokenizer,multilabel_binarizer)
        new_seuil=abs(seuil-prev_seuil)/2
        prev_seuil=seuil
        seuil=new_seuil
        seuil_table.append(seuil)
        acc_table.append(accuracy)
        prop_table.append(prop_classe)
        print(seuil,accuracy,prop_classe)
    return seuil_table,acc_table,prop_table

def find_seuil_top_k(model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    seuil_table=[]
    acc_table=[]
    prop_table=[]
    prev_seuil=0
    k=0
    accuracy=0
    prop_classe=0.5
    while(accuracy < 1 and prop_classe < 1 and k < len(multilabel_binarizer.classes_)) :
        (accuracy,prop_classe,precision) = test_pred_top_k(k,model,tokenizer,multilabel_binarizer)
        seuil_table.append(k)
        acc_table.append(accuracy)
        prop_table.append(prop_classe)
        print(k,accuracy,prop_classe)
        k=k+k//10+1
    return seuil_table,acc_table,prop_table
```

```python
def write_dep(th,rew,dep_file):
    with open(dep_file,"w+") as dep_f :
        dep_f.write("lemme :")
        for th in list(set(th_dep)) :
            dep_f.write(" "+th)
            
        dep_f.write("\nrewrite :")
        for rew in list(set(rew_dep)) :
            dep_f.write(" "+rew)

def dot_to_dep(dot_file,dep_file) :
    th_dep=[]
    rew_dep=[]
    ensure_dir(dep_file)
    with open(dot_file,"r") as dot_f :
        lines = dot_f.readlines()
        for line in lines :
            for word in types :
                if "\\\"Label"+word+"\\\"" in line :
                    rew_dep.append(word)
            for word in all_th :
                if "\\\"Label"+word+"\\\"" in line :
                    th_dep.append(word)
    write_dep(th,rew,dep_file)


            
def copy_dep(dep_file_from,dep_file_to) :
    ensure_dir(dep_file_to)
    copyfile(dep_file_from, dep_file_to)
    
    
def init_dep(dep_file) :
    ensure_dir(dep_file)
    with open(dep_file,"w+") as dep_f :
        dep_f.write("lemme :")
        dep_f.write("\nrewrite :")
```

```python
def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
        
def dep_to_zf(lemme,dep_file_name,zf_file_name) :
    
    dep_list=dep_to_list(dep_file_name)
    
    ensure_dir(zf_file_name)
    f=open(zf_file_name,"w+")
    f.write("include \"../tarski_term_def.zf\".\n")
    #if (set(dep_list) & set(grid_ok)) or lemme in grid_ok :
        #f.write("val O : point.\nval E : point.\nval E1 : point.\nassert ~Col O E E1.\n")
    for dep in dep_list :
        if dep in all_th :
            if dep==lemme :
                f.write("#include itself\n")
            else :
                f.write("#"+dep+"\n")
                f.write(all_th_declaration[dep])
        if dep in types :
            f.write(rewrite_declaration[dep])
            
    f.write("goal "+th_declaration_to_text(all_th_declaration[lemme]))
    f.close()
```

```python
def lemme_pred_to_dep(lemme,dep_file_name,seuil=0.05,model=model,tokenizer=t,max_length=max_length,multilabel_binarizer=multilabel_binarizer) :
    pred=lemme_pred(lemme,seuil,model=model,tokenizer=t,max_length=max_length)
    deps = multilabel_binarizer.inverse_transform(np.array([pred]))[0]
    th,rew=[]
    for dep in deps :
        if dep in all_th :
            th.append(dep)
        if dep in types :
            rew.append(dep)
    write_dep(th,rew,dep_file_name)
    
def th_string_pred_to_zf(th_string,zf_file_name,seuil=0.05,model=model,tokenizer=t,max_length=max_length,multilabel_binarizer=multilabel_binarizer) :
    pred=th_string_to_pred(th_string,seuil,model=model,tokenizer=t,max_length=max_length)
    deps = multilabel_binarizer.inverse_transform(np.array([pred]))[0]
    ensure_dir(zf_file_name)
    f=open(zf_file_name,"w+")
    f.write("include \"../tarski_term_def.zf\".\n")
    if (set(deps) & set(grid_ok)) :
        f.write("val O : point.\nval E : point.\nval E1 : point.\nassert ~Col O E E1.\n")
    for dep in deps :
        if dep in all_th :
            f.write("#"+dep+"\n")
            f.write(all_th_declaration[dep])
        if dep in types :
            f.write(rewrite_declaration[dep])
            
    f.write("goal "+th_declaration_to_text(th_string))
    f.close()
    
def lemme_pred_to_zf(lemme,zf_file_name=None,seuil=0.05,model=model,tokenizer=t,max_length=max_length,multilabel_binarizer=multilabel_binarizer) :
    pred=lemme_pred(lemme,seuil,model=model,tokenizer=t,max_length=max_length)
    deps = multilabel_binarizer.inverse_transform(np.array([pred]))[0]
    if zf_file_name is None :
        zf_file_name=zf_dir_path+get_chapter(lemme)+"/"+lemme+".zf"
    ensure_dir(zf_file_name)
    f=open(zf_file_name,"w+")
    f.write("include \"../tarski_term_def.zf\".\n")
    if (set(deps) & set(grid_ok)) :
        f.write("val O : point.\nval E : point.\nval E1 : point.\nassert ~Col O E E1.\n")
    for dep in deps :
        if dep in all_th :
            if dep==lemme :
                f.write("#include itself\n")
            else :
                f.write("#"+dep+"\n")
                f.write(all_th_declaration[dep])
        if dep in types :
            f.write(rewrite_declaration[dep])
            
    f.write("goal "+th_declaration_to_text(all_th_declaration[lemme]))
    f.close()
    
def th_string_to_zf_top_k(th_string,zf_file_name,k=5,model=model,tokenizer=t,max_length=max_length,multilabel_binarizer=multilabel_binarizer) :
    pred=th_string_to_pred_top_k(th_string,k,model=model,tokenizer=t,max_length=max_length)
    deps = multilabel_binarizer.inverse_transform(np.array([pred]))[0]
    ensure_dir(zf_file_name)
    f=open(zf_file_name,"w+")
    f.write("include \"../tarski_term_def.zf\".\n")
    if (set(deps) & set(grid_ok)) :
        f.write("val O : point.\nval E : point.\nval E1 : point.\nassert ~Col O E E1.\n")
    for dep in deps :
        if dep in all_th :
            f.write("#"+dep+"\n")
            f.write(all_th_declaration[dep])
        if dep in types :
            f.write(rewrite_declaration[dep])
            
    f.write("goal "+th_declaration_to_text(th_string))
    f.close()
    
def lemme_pred_to_zf_top_k(lemme,zf_file_name=None,k=5,model=model,tokenizer=t,max_length=max_length,multilabel_binarizer=multilabel_binarizer) :
    pred=lemme_pred_top_k(lemme,k,model=model,tokenizer=t,max_length=max_length)
    deps = multilabel_binarizer.inverse_transform(np.array([pred]))[0]
    if zf_file_name is None :
        zf_file_name=zf_dir_path+get_chapter(lemme)+"/"+lemme+".zf"
    ensure_dir(zf_file_name)
    f=open(zf_file_name,"w+")
    f.write("include \"../tarski_term_def.zf\".\n")
    if (set(deps) & set(grid_ok)) :
        f.write("val O : point.\nval E : point.\nval E1 : point.\nassert ~Col O E E1.\n")
    for dep in deps :
        if dep in all_th :
            if dep==lemme :
                f.write("#include itself\n")
            else :
                f.write("#"+dep+"\n")
                f.write(all_th_declaration[dep])
        if dep in types :
            f.write(rewrite_declaration[dep])
            
    f.write("goal "+th_declaration_to_text(all_th_declaration[lemme]))
    f.close()
```

```python


n_jobsZ=6

semZ = Semaphore(n_jobsZ)
lockZ = RLock()

zipper_exec_time=dict()

class Zipperposition(Thread):

    """Thread chargé simplement d'afficher une lettre dans la console."""

    def __init__(self, zf_dir,chapter,lemme):
        Thread.__init__(self)
        self.zf_dir=zf_dir
        self.chapter=chapter
        self.lemme=lemme

    def run(self):
        """Code à exécuter pendant l'exécution du thread."""
        global zipper_exec_time
        t=time.time()
        lemme_file=self.zf_dir+self.chapter+"/"+self.lemme
        if os.path.isfile(lemme_file+".zf") :
            res=os.popen("zipperposition -o none --mem-limit 4000 --timeout 30 --dot "+lemme_file+".dot "+lemme_file+".zf").read()
            with lockZ :
                if res.endswith("output end Refutation\n") :
                    print(self.lemme+" : Valid")
                    zipper_exec_time[self.chapter][self.lemme]=(time.time()-t)
                else:
                    print(self.lemme+" : Error")
                    zipper_exec_time[self.chapter][self.lemme]=res
        else :
            with lockZ :
                print("FileNotFoundError : "+lemme_file+".zf")
            
```

```python
def get_zipper_stat(zipper_exec_time):
    done=[]
    notdone=[]
    gaveup=[]
    ressourceout=[]
    timeout=[]
    for chapter in zipper_exec_time :
        for lemme in zipper_exec_time[chapter] :
            if isinstance(zipper_exec_time[chapter][lemme],float) :
                done.append(lemme)
            else :
                if "GaveUp" in zipper_exec_time[chapter][lemme] :
                    gaveup.append(lemme)
                else :
                    if "for" in zipper_exec_time[chapter][lemme] :
                        timeout.append(lemme)
                    else :
                        ressourceout.append(lemme)
                notdone.append(lemme)
    return (done,notdone,gaveup,ressourceout,timeout)
```

```python
#grid_ok=[]

def zipper_stat(done,notdone,gaveup,ressourceout,timeout):
    res = str(len(done)/(len(done)+len(notdone))*100)+"% good to go = "+str(len(notdone))
    if len(notdone) > 0 :
        res+=" GaveUp = "+str(len(gaveup)/len(notdone)*100)+" TimeOut = "+str(len(timeout)/len(notdone)*100)+" RessourceOut = "+str(len(ressourceout)/len(notdone)*100)+" size = "+str(len(done)+len(notdone))
    return res
```

```python
def tactics(list_of_lemme=th_valid,seuil=0.9, nb_iter=100) :
    
    if not(list_of_lemme and nb_iter > 0 and seuil > 0.00001) : #possède au moins un élément
        return []
    
    global zipper_exec_time
    print("size_of_echantillon =",len(list_of_lemme))
    print("starting")
    
    zipper_exec_time=dict()
    
    print("create files")
    
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme in list_of_lemme :
                lemme_pred_to_zf(lemme,seuil=seuil)
                
    print("file created")
                
    threads=[]

    for chapter in geocoq :

        zipper_exec_time[chapter]=dict()

        for lemme in geocoq[chapter] :
            if lemme in list_of_lemme :
                threads.append(Zipperposition(zf_dir_path,chapter,lemme))

    for i in range(len(threads)):
        with semZ :
            threads[i].start()

    for i in range(len(threads)):
        threads[i].join()
    
    done,notdone,gaveup,ressourceout,timeout = get_zipper_stat(zipper_exec_time)
    
    print("iter remaining : "+str(nb_iter))
    print(zipper_stat(done,notdone,gaveup,ressourceout,timeout))

    data_list = tactics(list_of_lemme=notdone,seuil=seuil/1.2,nb_iter=nb_iter-1)
    return [[seuil,done,notdone,gaveup,ressourceout,timeout]]+data_list
```

```python
def tactics_top_k(list_of_lemme=th_valid,k=0, nb_iter=100) :
    
    if not(list_of_lemme and nb_iter > 0 and k < 180) : #possède au moins un élément
        return []
    
    global zipper_exec_time
    print("size_of_echantillon =",len(list_of_lemme))
    print("starting")
    
    zipper_exec_time=dict()
    
    print("create files")
    
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme in list_of_lemme :
                lemme_pred_to_zf_top_k(lemme,k=k)
                
    print("file created")
                
    threads=[]

    for chapter in geocoq :

        zipper_exec_time[chapter]=dict()

        for lemme in geocoq[chapter] :
            if lemme in list_of_lemme :
                threads.append(Zipperposition(zf_dir_path,chapter,lemme))

    for i in range(len(threads)):
        with semZ :
            threads[i].start()

    for i in range(len(threads)):
        threads[i].join()
    
    done,notdone,gaveup,ressourceout,timeout = get_zipper_stat(zipper_exec_time)
    
    print("k="+str(k))
    print(zipper_stat(done,notdone,gaveup,ressourceout,timeout))

    data_list = tactics_top_k(list_of_lemme=notdone,k=k+k//10+1,nb_iter=nb_iter-1)
    return [[k,done,notdone,gaveup,ressourceout,timeout]]+data_list
```

```python
def test_dep(dep_dir,zf_test_dir) :
    for chapter in geocoq :

        dep_path=dep_dir+chapter+"/"
        zf_path=zf_test_dir+chapter+"/"

        for lemme_name in geocoq[chapter] :

            ensure_dir(zf_path+lemme_name+".zf")
            

            dep_file=open(dep_path+lemme_name+".dep","r")

            for line in dep_file.readlines() :
                if line.startswith("lemme :") :
                    inter_lemmes = line.split()[2:]
                if line.startswith("rewrite :") :
                    rewrites = line.split()[2:]

            dep_file.close()
            
            
            zf_file=open(zf_path+lemme_name+".zf","w+")
            
            zf_file.write("include \"../tarski_term_def.zf\".\n")
            if (lemme_name in grid_ok) :
                zf_file.write("val O : point.\nval E : point.\nval E1 : point.\nassert ~Col O E E1.\n")
            for dep in inter_lemmes :
                if dep==lemme :
                    zf_file.write("#include itself\n")
                else :
                    zf_file.write("#"+dep+"\n")
                    zf_file.write(all_th_declaration[dep])
                    
            for dep in rewrites :
                if dep in zf_cant_handle_commat :
                    dep=dep.replace('\'','')
                zf_file.write(rewrite_declaration[dep])

            zf_file.write("goal "+th_declaration_to_text(all_th_declaration[lemme_name]))
            zf_file.close()
```
