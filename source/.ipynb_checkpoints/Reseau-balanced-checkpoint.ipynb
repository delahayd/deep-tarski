```python
zf_dir_path="./zf_init/"
dep_dir_path = "./dependencies_init/"

import re
import numpy as np

def get_rewrites_definitions(tarski_def_file_path) :
    rewrite = dict()

    tarski_def_file=open(tarski_def_file_path,"r")

    for line in tarski_def_file.readlines():
        if line.startswith('rewrite') :
            words=re.findall(r"[\w']+", line)
            for w in words :
                if w in types :
                    type_rew=w
                    break
            try:
                rewrite[type_rew]=line
            except:
                print(line)
                break
                
    tarski_def_file.close()
    return rewrite

def get_all_th_definitions(all_th_file_path) :
    th_decl = dict()
    all_th_file=open(all_th_file_path,"r")
    
    lines = all_th_file.readlines()
    
    for i in range(len(lines)) :
        if lines[i].startswith("#") :
            name_size = len(lines[i])
            th_decl[lines[i][1:name_size-1]]=lines[i+1]
    return th_decl
```

```python
def dep_to_list(dep_file_path) :
    res=[]
    dep_file=open(dep_file_path,"r")
    for line in dep_file.readlines() :
        line=line.replace("\n","")
        line=line.replace("\r","")
        if line.startswith("lemme :") :
            for w in line.split(" ") :
                if w in all_th :
                    res.append(w)
        if line.startswith("rewrite :") :
            for w in line.split(" ") :
                if w in types :
                    res.append(w)
    dep_file.close()
    return res

def dep_to_categorical(dep_file_path) :
    vector=[0 for i in range(len(index))]
    dep_file=open(dep_file_path,"r")
    for line in dep_file.readlines() :
        line=line.replace("\n","")
        line=line.replace("\r","")
        if line.startswith("lemme :") :
            for w in line.split(" ") :
                if w in all_th :
                    vector[index.index(w)]=1
        if line.startswith("rewrite :") :
            for w in line.split(" ") :
                if w in types :
                    vector[index.index(w)]=1
    dep_file.close()
    return vector


def categorical_to_dep(y,prop_seuil=0,nb_output=0):
    res=[]
    seuil=prop_seuil*max(y)
    for i in range(len(y)) :
        if y[i] >= seuil or seuil==0 :
            res.append((y[i],index[i]))
    if nb_output==0 :
        return sorted(res,reverse=True)
    else :
        return sorted(res,reverse=True)[:nb_output]
    
    
def get_name_of_dep_output(dep_output):
    return list(list(zip(*dep_output))[1])
    
def get_categorical_dep_output(y,dep_names) :
    res=[]
    cat_dep=categorical_to_dep(y)
    for dep in cat_dep :
        if dep[1] in dep_names :
            res.append(dep)
    return sorted(res,reverse=True)
```

```python
def extract_value(y) :
    return [val for (val,classe) in y]

def extract_balanced_classes(y) :
    return np.concatenate([classe for (val,classe) in y])

def occurence_of_classes(y) :
    res=[0 for i in range(len(y[0]))]
    for vect in y :
        for i in range(len(vect)) :
            if vect[i]==1 :
                res[i]+=1
    return res

def get_unused_classes(y) :
    oc_cl = occurence_of_classes(y)
    categoricals=categorical_to_dep(oc_cl)
    return [key for (value,key) in categoricals if value==0]

def get_used_classes(y) :
    oc_cl = occurence_of_classes(y)
    categoricals=categorical_to_dep(oc_cl)
    return [(value,key) for (value,key) in categoricals if value!=0]

def concat_classes(c0,c1) :
    return (c0[0]+c1[0],c0[1]+c1[1])


def create_balanced_classes(y, prop_variance_seuil=0.1, nb_class_min=20, max_iter=100) :
    used_classes = [(val,[classe]) for (val,classe) in get_used_classes(y)]
    nb_iter=0
    conv=False
    while (not(conv) and len(used_classes) > nb_class_min and (nb_iter < max_iter)):
        res=[]
        
        actual_len=len(used_classes)
        
        val_max = np.max(extract_value(used_classes))
        
        i=0
        
        while i < len(used_classes) :
            found_better=False
            if np.var(extract_value(used_classes[:i+1])) > val_max*prop_variance_seuil :
                j=len(used_classes)-1
                while not(found_better) and j>i :
                    #if np.var(extract_value(used_classes[:i+1])) > np.var(extract_value(used_classes[:i]+[concat_classes(used_classes[i],used_classes[j])])) :
                    if abs(used_classes[i][0]-val_max) > abs((used_classes[i][0]+used_classes[j][0])-val_max) :
                    #if abs(val_max-np.mean(extract_value(used_classes[:i+1]))) >= abs(val_max-np.var(extract_value(used_classes[:i]+[concat_classes(used_classes[i+1],used_classes[j])]))) :
                        res.append(concat_classes(used_classes[i],used_classes[j]))
                        del(used_classes[j])
                        found_better=True
                    j-=1
            if not(found_better) :
                res.append(used_classes[i])
            i+=1
                    
        used_classes=sorted(res,reverse=True)[:]
        nb_iter+=1
        if actual_len == len(used_classes) :
            conv=True
    return used_classes

        
def get_diff_prop_occurence(y) :
    return np.max(y)/np.min(y)
```

```python
def balanced_dep_to_categorical(dep_file_path,balanced_classes=balanced_classes) :
    vector=[0 for i in range(len(balanced_classes))]
    dep_file=open(dep_file_path,"r")
    for line in dep_file.readlines() :
        line=line.replace("\n","")
        line=line.replace("\r","")
        if line.startswith("lemme :") :
            for w in line.split(" ") :
                for i in range(len(balanced_classes)) :
                    if w in balanced_classes[i] :
                        vector[i]=1
        if line.startswith("rewrite :") :
            for w in line.split(" ") :
                for i in range(len(balanced_classes)) :
                    if w in balanced_classes[i] :
                        vector[i]=1
    dep_file.close()
    return vector

def balanced_categorical_to_dep(y,prop_seuil=0,nb_output=0,balanced_classes=balanced_classes):
    res=[]
    seuil=prop_seuil*np.max(y)
    for i in range(len(y)) :
        if y[i] >= seuil or seuil==0 :
            res.append((y[i],balanced_classes[i]))
    if nb_output==0 :
        return sorted(res,reverse=True)
    else :
        return sorted(res,reverse=True)[:nb_output]
    
def get_balanced_categorical_dep_output(y,dep_names,balanced_classes=balanced_classes) :
    res=[]
    b_cat_dep=balanced_categorical_to_dep(y)
    for deps in b_cat_dep :
        for dep in deps[1] :
            if dep in dep_names :
                res.append(deps)
    return sorted(res,reverse=True)
```

```python
def get_chapter(lemme_to_find) :
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme==lemme_to_find :
                return chapter
        
def balanced_lemme_pred(lemme,seuil=0.05,model=model,tokenizer=t,max_length=max_length,balanced_classes=balanced_classes) :
    x = all_th_declaration[lemme].replace("assert ","").replace("\n","")
    x = pretraitement(np.array([x]))
    
    encoded_docs = tokenizer.texts_to_sequences(x)
    
    padded_docs = prepro.sequence.pad_sequences(encoded_docs, maxlen=max_length, padding='post')
    
    balanced_y_pred = model.predict(padded_docs)[0]
    y_pred=balanced_pred_to_pred(balanced_y_pred)    
                        
    max_pred_value = np.max(y_pred)
    y_pred = np.array([0 if val < max_pred_value*seuil else 1 for val in y_pred])
    
    return y_pred

def is_good_pred(lemme,seuil=0.05,model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    pred_y = balanced_lemme_pred(lemme,seuil,model,tokenizer)
    
    pred_y_ones = np.where(pred_y==1)[0]
    
    
    labeled_y = dep_to_list(dep_dir_path+"/"+get_chapter(lemme)+"/"+lemme+".dep")
    y = multilabel_binarizer.transform(np.array([labeled_y]))
    
    y_ones = np.where(y[0]==1)[0]
    
    compare = [1 if val in pred_y_ones else 0 for val in y_ones]
    
    if len(compare)==0 :
        compare=[0]
    
    return (np.average(compare),len(pred_y_ones)/len(pred_y))


print(is_good_pred('cong_reflexivity',seuil=0.3))
```

```python
def test_pred(seuil=0.05,model=model,tokenizer=t,multilabel_binarizer=multilabel_binarizer) :
    sum_res=0
    size=0
    total_precision=None
    total_prop_classe=None
    for chapter in geocoq :
        for lemme in geocoq[chapter] :
            if lemme in th_valid :
                (res,prop_classe) = is_good_pred(lemme,seuil,model,tokenizer,multilabel_binarizer)
                if res==1 :
                    sum_res+=1
                else :
                    if total_precision is None :
                        total_precision=res
                    else :
                        total_precision+=res
                if total_prop_classe is None :
                    total_prop_classe=prop_classe
                else :
                    total_prop_classe+=prop_classe
                size+=1
    if total_precision is None :
        total_precision=0
    if total_prop_classe is None :
        total_prop_classe=0
    return (sum_res/size,total_prop_classe/size,total_precision/size)
```

```python
import os
import time
from threading import Thread, Semaphore, RLock

n_jobsZ=6

semZ = Semaphore(n_jobsZ)
lockZ = RLock()

zipper_exec_time=dict()

class Zipperposition(Thread):

    """Thread chargÃ© simplement d'afficher une lettre dans la console."""

    def __init__(self, zf_dir,chapter,lemme):
        Thread.__init__(self)
        self.zf_dir=zf_dir
        self.chapter=chapter
        self.lemme=lemme

    def run(self):
        """Code Ã  exÃ©cuter pendant l'exÃ©cution du thread."""
        global zipper_exec_time
        t=time.time()
        lemme_file=self.zf_dir+self.chapter+"/"+self.lemme+".zf"
        if os.path.isfile(lemme_file) :
            res=os.popen("zipperposition -o none --mem-limit 4000 --timeout 30 "+lemme_file).read()
            with lockZ :
                if res.endswith("output end Refutation\n") :
                    print(self.lemme+" : Valid")
                    zipper_exec_time[self.chapter][self.lemme]=(time.time()-t)
                else:
                    print(self.lemme+" : Error")
                    zipper_exec_time[self.chapter][self.lemme]=res
        else :
            with lockZ :
                print("FileNotFoundError : "+lemme_file)
            
```

```python
def get_zipper_stat(zipper_exec_time):
    done=[]
    notdone=[]
    gaveup=[]
    ressourceout=[]
    timeout=[]
    for chapter in zipper_exec_time :
        for lemme in zipper_exec_time[chapter] :
            if isinstance(zipper_exec_time[chapter][lemme],float) :
                done.append(lemme)
            else :
                if "GaveUp" in zipper_exec_time[chapter][lemme] :
                    gaveup.append(lemme)
                else :
                    if "for" in zipper_exec_time[chapter][lemme] :
                        timeout.append(lemme)
                    else :
                        ressourceout.append(lemme)
                notdone.append(lemme)
    return (done,notdone,gaveup,ressourceout,timeout)
```

```python
#grid_ok=[]

def zipper_stat(done,notdone,gaveup,ressourceout,timeout):
    res = str(len(done)/(len(done)+len(notdone))*100)+"% good to go = "+str(len(notdone))
    if len(notdone) > 0 :
        res+=" GaveUp = "+str(len(gaveup)/len(notdone)*100)+" TimeOut = "+str(len(timeout)/len(notdone)*100)+" RessourceOut = "+str(len(ressourceout)/len(notdone)*100)+" size = "+str(len(done)+len(notdone))
    return res
```
